```python
import glob
import pandas as pd

# Get a list of all CSV files in the directory
csv_files = glob.glob('process-by-station/*.csv')

# Read each CSV file into a DataFrame and store them in an array
dataframes = [pd.read_csv(file) for file in csv_files]

# Display the array of DataFrames
dataframes
```

```python
filtered_dataframes = [df[df['REPORT_TYPE'].isin(['FM-12', 'FM-15', 'FM-16'])] for df in dataframes]

# Display the filtered array of DataFrames
filtered_dataframes
```

```python
filtered_dataframes_trimed = [df.loc[:, df.columns.str.startswith('Hourly') | df.columns.isin(['STATION', 'REPORT_TYPE', 'DATE'])] for df in filtered_dataframes]

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
from collections import Counter

def most_common_value(series):
    return series.mode().iloc[0] if not series.mode().empty else None

def aggregate_dataframe(df):
    return df.groupby(['STATION', 'DATE']).agg(
        {col: lambda x: x.mean(skipna=True) if df[col].dtype in ['float64', 'int64'] else most_common_value for col in df.columns if col not in ['STATION', 'DATE', 'REPORT_TYPE']}
    ).reset_index()

aggregated_dataframes = [aggregate_dataframe(df) for df in filtered_dataframes_trimed]

# Display the aggregated array of DataFrames
aggregated_dataframes
```
