```python
import glob
import pandas as pd

# Load all CSV files from Raw 2023 folder
raw_2023_files = glob.glob('Raw 2023/*.csv')
raw_2023_data = [pd.read_csv(file) for file in raw_2023_files]

# Load all CSV files from Raw 2024 folder
raw_2024_files = glob.glob('Raw 2024/*.csv')
raw_2024_data = [pd.read_csv(file) for file in raw_2024_files]

# Combine all data into a single DataFrame for each year
raw_2023_combined = pd.concat(raw_2023_data, ignore_index=True)
raw_2024_combined = pd.concat(raw_2024_data, ignore_index=True)

# Combine raw_2023_combined and raw_2024_combined
raw_combined = pd.concat([raw_2023_combined, raw_2024_combined], ignore_index=True)
```

```python
sort_station_combined = raw_combined.groupby(['STATION'])
sod_combined = sort_station_combined.apply(lambda x: x[x['REPORT_TYPE'].str.strip() == 'SOD'])

print(sod_combined)
```

```python
import os

# Create the folder if it doesn't exist
output_folder = 'process-by-date'
os.makedirs(output_folder, exist_ok=True)

for station, group in sod_combined.groupby(level='STATION'):
    output_file = os.path.join(output_folder, f'{station}.csv')
    group.to_csv(output_file, index=False)
```

```python
import pandas as pd
import glob

# Get all CSV files in the process-by-date folder
csv_files = glob.glob('process-by-date/*.csv')

for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Convert the 'DATE' column to datetime
    df['DATE'] = pd.to_datetime(df['DATE'])
    
    # Sort the dataframe by the 'DATE' column
    df_sorted = df.sort_values(by='DATE')
    
    # Save the sorted dataframe back to the CSV file
    df_sorted.to_csv(file, index=False)
```

```python
for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Get the number of entries
    num_entries = len(df)
    
    # Get the number of missing values
    num_missing_values = df.isnull().sum().sum()
    
    print(f'File: {file}')
    print(f'Number of entries: {num_entries}')
    print(f'Number of missing values: {num_missing_values}')
    print('---')
```

```python

# Get the stations with 600 and above entries from the process-by-date folder
stations_with_600_and_above_files = []
total_entries = 0
for file in csv_files:
    df = pd.read_csv(file)
    if len(df) >= 600:
        stations_with_600_and_above_files.append(file)
        total_entries += len(df)

print(stations_with_600_and_above_files)
print(len(stations_with_600_and_above_files)) 
print(total_entries)
```
