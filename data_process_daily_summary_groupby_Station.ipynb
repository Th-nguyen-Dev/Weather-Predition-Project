```python
import glob
import pandas as pd

# Get a list of all CSV files in the directory
csv_files = glob.glob('process-by-station/*.csv')

# Read each CSV file into a DataFrame and store them in an array
dataframes = [pd.read_csv(file) for file in csv_files]

# Display the array of DataFrames
dataframes
```

```python
filtered_dataframes = [df[df['REPORT_TYPE'].isin(['FM-12', 'FM-15', 'FM-16'])] for df in dataframes]

# Display the filtered array of DataFrames
filtered_dataframes
```

```python
filtered_dataframes_trimed = [df.loc[:, df.columns.str.startswith('Hourly') | df.columns.isin(['STATION', 'REPORT_TYPE', 'DATE'])] for df in filtered_dataframes]

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
def clean_and_convert_to_float(series):
    return pd.to_numeric(series.astype(str).str.extract(r'(\d+\.?\d*)')[0], errors='coerce')

for df in filtered_dataframes_trimed:
    if "HourlyDewPointTemperature" in df.columns:
        df["HourlyDewPointTemperature"] = clean_and_convert_to_float(df["HourlyDewPointTemperature"])
    if "HourlyDryBulbTemperature" in df.columns:
        df["HourlyDryBulbTemperature"] = clean_and_convert_to_float(df["HourlyDryBulbTemperature"])
    if "HourlyWetBulbTemperature" in df.columns:
        df["HourlyWetBulbTemperature"] = clean_and_convert_to_float(df["HourlyWetBulbTemperature"])
    if "HourlyStationPressure" in df.columns:
        df["HourlyStationPressure"] = clean_and_convert_to_float(df["HourlyStationPressure"])
    if "HourlyAltimeterSetting" in df.columns:
        df["HourlyAltimeterSetting"] = clean_and_convert_to_float(df["HourlyAltimeterSetting"])

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
def clean_and_convert_to_float_precip(series):
    series = series.replace('T', 0.005)
    return pd.to_numeric(series.astype(str).str.extract(r'(\d+\.?\d*)')[0], errors='coerce')

for df in filtered_dataframes_trimed:
    if "HourlyPrecipitation" in df.columns:
        df["HourlyPrecipitation"] = clean_and_convert_to_float_precip(df["HourlyPrecipitation"])

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
for df in filtered_dataframes_trimed:
    if "HourlyWindSpeed" in df.columns:
        df["HourlyWindSpeed"] = clean_and_convert_to_float(df["HourlyWindSpeed"])

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
for df in filtered_dataframes_trimed:
    if 'HourlyVisibility' in df.columns:
        df.drop(columns=['HourlyVisibility'], inplace=True)

# Display the updated array of DataFrames
filtered_dataframes_trimed
```

```python
from collections import Counter

# Show each column type for the first dataframe in the list
print(filtered_dataframes_trimed[0].dtypes)
for col in filtered_dataframes_trimed[0].select_dtypes(include=['object']).columns:
    if col not in ['DATE', 'REPORT_TYPE', 'HourlyPresentWeatherType']:
        print(f"Unique values in column {col}: {filtered_dataframes_trimed[0][col].unique()}")

def most_common_value(series):
    return series.mode().iloc[0] if not series.mode().empty else None

def aggregate_dataframe(df):
    df['DATE'] = pd.to_datetime(df['DATE']).dt.date
    aggregation_functions = {}
    for col in df.columns:
        if col not in ['STATION', 'DATE', 'REPORT_TYPE']:
            if pd.api.types.is_numeric_dtype(df[col]):
                aggregation_functions[col] = lambda x: x.mean(skipna=True)
            else:
                aggregation_functions[col] = most_common_value
    return df.groupby(['STATION', 'DATE']).agg(aggregation_functions).reset_index()

aggregated_dataframes = [aggregate_dataframe(df) for df in filtered_dataframes_trimed]

# Display the aggregated array of DataFrames
aggregated_dataframes
```

```python
def print_station_info(dataframes):
    for df in dataframes:
        if not df.empty:
            station_id = df['STATION'].iloc[0]
            num_entries = len(df)
            print(f'Station ID: {station_id}, Number of Entries: {num_entries}')

print_station_info(aggregated_dataframes)
```

```python
filtered_aggregated_dataframes = [df for df in aggregated_dataframes if len(df) >= 900]

# Display the filtered array of DataFrames
print_station_info(filtered_aggregated_dataframes)
```

```python
def ensure_entries(df):
    # Generate a date range from January 1 to October 27
    date_range = pd.date_range(start='2022-01-01', end='2024-10-27')
    all_dates_df = pd.DataFrame(date_range, columns=['DATE'])
    all_dates_df['STATION'] = df['STATION'].iloc[0]
    
    # Ensure the 'DATE' column in df is of type datetime
    df['DATE'] = pd.to_datetime(df['DATE'])
    
    # Merge with the original dataframe to find missing dates
    merged_df = all_dates_df.merge(df, on=['DATE', 'STATION'], how='left')
    
    # Backfill missing entries
    merged_df.fillna(method='bfill', inplace=True)
    merged_df.fillna(method='ffill', inplace=True)
    
    return merged_df

filtered_aggregated_dataframes_filled = [ensure_entries(df) for df in filtered_aggregated_dataframes]
```

```python
for df in filtered_aggregated_dataframes_filled:
    if 'HourlyWindGustSpeed' in df.columns:
        df.drop(columns=['HourlyWindGustSpeed'], inplace=True)

# Display the updated array of DataFrames
filtered_aggregated_dataframes_filled
```

```python
def count_nulls(df):
    # Replace '*' with NaN
    df.replace('*', pd.NA, inplace=True)
    # Count null values for each column
    null_counts = df.isna().sum()
    return null_counts
def count_nulls_per_station(dataframes):
    # Apply the function to each dataframe in the list
    null_counts_per_station = {df['STATION'].iloc[0]: count_nulls(df) for df in dataframes if not df.empty}

    # Display the null counts for each station
    for station, null_counts in null_counts_per_station.items():
        print(f'Station ID: {station}')
        print(null_counts)
        print()

# Call the function with the aggregated_dataframes
count_nulls_per_station(filtered_aggregated_dataframes_filled)
```

```python
def discard_null_dataframes(dataframes):
    return [df for df in dataframes if count_nulls(df).sum() == 0]

filtered_aggregated_dataframes_no_nulls = discard_null_dataframes(filtered_aggregated_dataframes_filled)

# Display the filtered array of DataFrames
print_station_info(filtered_aggregated_dataframes_no_nulls)
```

```python
count_nulls_per_station(filtered_aggregated_dataframes_no_nulls)
```

```python
import os

# Create the directory if it doesn't exist
output_dir = 'draft-final-data-2'
os.makedirs(output_dir, exist_ok=True)
print_station_info(filtered_aggregated_dataframes_no_nulls)
# Export each dataframe to a CSV file with the station ID as the filename
for df in filtered_aggregated_dataframes_no_nulls:
    df.columns = [col.replace('Hourly', 'Daily') if col.startswith('Hourly') else col for col in df.columns]
    if not df.empty:
        station_id = df['STATION'].iloc[0]
        output_path = os.path.join(output_dir, f'{station_id}.csv')
        df.to_csv(output_path, index=False)
```
