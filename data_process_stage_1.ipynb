```python
import glob
import pandas as pd
import os
import shutil

# Get all CSV files from raw 2020 to raw 2024 directories
csv_files_2020 = glob.glob('./Raw 2020/*.csv')
csv_files_2021 = glob.glob('./Raw 2021/*.csv')
csv_files_2022 = glob.glob('./Raw 2022/*.csv')
csv_files_2023 = glob.glob('./Raw 2023/*.csv')
csv_files_2024 = glob.glob('./Raw 2024/*.csv')

# Combine the lists
all_csv_files = csv_files_2023 + csv_files_2024 + csv_files_2022 + csv_files_2021 + csv_files_2020

# Print the list of all CSV files
print(all_csv_files)

# Initialize an empty dataframe
combined_df = pd.DataFrame()

# Read all CSV files in chunks and concatenate them
for file in all_csv_files:
    for chunk in pd.read_csv(file, chunksize=100000):
        combined_df = pd.concat([combined_df, chunk])

# Convert 'DATE' column to datetime
combined_df['DATE'] = pd.to_datetime(combined_df['DATE'])

# Group by 'STATION' and sort each group by 'DATE'
grouped_df = combined_df.groupby('STATION').apply(lambda x: x.sort_values('DATE')).reset_index(drop=True)

# Delete folder if it already exists
output_dir = './process-by-station'
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

# Create directory
os.makedirs(output_dir, exist_ok=True)

# Iterate over each unique station and save to CSV
unique_stations = grouped_df['STATION'].unique()
for station in unique_stations:
    station_df = grouped_df[grouped_df['STATION'] == station]
    output_file = os.path.join(output_dir, f'{station}.csv')
    station_df.to_csv(output_file, index=False)

print(f"CSV files have been saved to {output_dir}")

# Print the number of rows in the combined dataframe
print(f"Total number of rows in the combined dataframe: {combined_df.shape[0]}")

# Print the number of rows for each station
for station in unique_stations:
    station_df = grouped_df[grouped_df['STATION'] == station]
    print(f"Number of rows for station {station}: {station_df.shape[0]}")
```
